```python
import json
```

```python
data = open('./p20.txt', 'r')
```

```python
# lines = data.readlines()
```

```python
# define dictionaries for replacement info
replace_bus = {}
bus_distance = {}
charge_sequence = {}
charge_stations = {}
equity_value = ""
```

```python
for line in data:
    # process the environmental equity line
    if "value" in line:
        equity_line = line.split()
        equity_value = float(equity_line[4])
#         print(equity_value)
    else:
        tokens = line.split();
        if len(tokens) != 0:
            key, value = tokens[0], tokens[1]
            
            # process the buses to be replaced
            if key[0] == 'Z':
                bus = int(key[1:])
#                 print(bus)
                replace_bus[bus] = int(value)
    
            # process the distance traveled
            elif key[0] == 'm':
                bus_end = key.index('_')
                bus = int(key[1:bus_end])
                if bus not in bus_distance:
                    bus_distances = []
                    bus_distances.append(float(value))
                    bus_distance[bus] = bus_distances
                else:
                    bus_distances = bus_distance.get(bus)
                    bus_distances.append(float(value))
                    bus_distance[bus] = bus_distances
                    
            # process the bus charge sequence 
            elif key[0] == 'X':
                bus_end = key.index('_')
                bus = int(key[1:bus_end])
                if bus not in charge_sequence:
                    bus_charge = []
                    bus_charge.append(int(value))
                    charge_sequence[bus] = bus_charge
                else:
                    bus_charge = charge_sequence.get(bus)
                    bus_charge.append(int(value))
                    charge_sequence[bus] = bus_charge
            # process the number of charge stations
            elif key[0] == 'Y':
                stop_id = int(key[1:])
                charge_stations[stop_id] = int(value)
                
                
# final result 
processed_data = {"Environmental Equity": equity_value,
                 "Replace Buses": replace_bus,
                 "Bus Distance": bus_distance,
                 "Charge Sequence": charge_sequence,
                 "Charge Stations": charge_stations}

new_data = json.dumps(processed_data, indent = 4)
print(new_data)
            
                

# print(f"The equity value is: {equity_value}")
# print(f"The bus replacements: {replace_bus}")
# print(f"The bus distances: {bus_distance}")
# print(f"The charge sequence: {charge_sequence}")
# print(f"The charge stations: {charge_stations}")
# print(processed_data)

```
